#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

```python
plt.style.use('ggplot')
plt.figure(figsize=(15, 15))
fig, ax = plt.subplots(2)

ax[0].title.set_text('Lineal')
ax[0].set_xticklabels([]) #sin etiquetas
ax[0].plot(x, y)

ax[1].title.set_text('Cuadr√°tica')
ax[1].scatter(x, z)
plt.xticks(rotation=90)

plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)`¬†in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
z = np.exp(x)
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(15, 10))

plt.plot(x, x*2, color='r', linewidth=4)
plt.plot(x, z, color='g', linewidth=4)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_title('Logaritmic');
```

#### In the vehicles data set that you have downloaded, use the vehicles.csv file. In this exercise we will conduct some exploratory data analysis using one plot each of scatter plot, box plot, histogram, and bar chart. 

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
data = pd.read_csv('vehicles.csv')
```

```python
data.head()
```

```python
x = data['Combined MPG']
y = data['Highway MPG']

plt.figure(figsize=(10,8))

plt.scatter(x, y, alpha=0.5, norm=0.5, color='y')
plt.title('Combined MPG Vs Highway MPG', fontsize=25)
plt.ylabel('Highway MPG', fontsize=18)
plt.xlabel('Combined MPG', fontsize=18);
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
plt.figure(figsize=(10,6))
plt.boxplot(data['CO2 Emission Grams/Mile'], labels=['CO2 Emission Grams/Mile'], showmeans=True);
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
plt.figure(figsize=(10,6))
plt.hist(data['Fuel Barrels/Year'], bins=60, color='r')
plt.legend(['Fuel Barrels/Year']);
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
df= data[['Fuel Type', 'City MPG']].groupby('Fuel Type', as_index=False).mean('City MPG')
x = df['Fuel Type']
y = df['City MPG']
```

```python
plt.figure(figsize=(10,6))
plt.bar(x, height=y, color='c')
plt.xticks(rotation=90);
```
